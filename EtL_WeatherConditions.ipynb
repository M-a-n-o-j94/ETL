```python
def current_weather(city):
    key="b6729a0fec15473083d42727242008"

    url=f"http://api.weatherapi.com/v1/current.json?key={key}&q={city}&aqi=no"
    response=requests.get(url).json()
    df=pd.DataFrame(response)
    return df["location"].combine_first(df["current"])
```

```python
cities=['New York', 'Los Angeles', 'Chicago', 'Miami', 'London']
li=[]
for i in cities:
  li.append(current_weather(i))


```
